```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#data load
df=pd.read_csv("customer_clean_data.csv")
df
```

## 1. **Gender Distribution by Country**: 
### What is the ratio of Male and Female customers in Pakistan and other countries? (Stacked Bar Chart)

```python

dp=pd.crosstab(df["Country"],df["Gender"])
dp.plot(kind='bar',stacked=True,figsize=(8,6))
plt.title("Gender Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Number of people")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

## 2. **Spending Patterns by Membership**: 
### Is there a significant difference in `TotalSpent` between 'Gold', 'Silver', and 'Bronze' membership customers? (Box Plot)


```python
df["Membership"].replace("gold", "Gold",inplace=True)
```

```python
sns.boxenplot(x=df["Membership"],y=df["TotalSpent"],palette="viridis")
plt.Figure(figsize=(8,6))

plt.title("Spending Patterns by Membership")
plt.xlabel("Membership")
plt.ylabel("Spending")
plt.show()
```

## 3. **Age vs. Orders Correlation**: 
### Is there a clear change in the total number of `Orders` as customer Age increases? (Scatter Plot)


```python
plt.figure(figsize=(8,6))
plt.scatter(df["Age"],df["Orders"],alpha=0.6, color='blue')
plt.xlabel("Age")
plt.ylabel("Order")
plt.grid(True,linestyle='--',alpha=0.7)
plt.title("Age vs Order Correlation")
plt.show()
```

## 4. **Payment Method Preference by Gender**:
### Which payment methods (Crypto, Card, Cash) do Male and Female customers use most frequently? (Grouped Bar Chart)


```python
print(df.columns)
```

```python
plt.figure(figsize=(8,6))
payment_data=pd.crosstab(df["Gender"],df["PaymentMethod"])
payment_data.plot(kind='bar')
plt.title("Payment Method Preference by Gender")
plt.xlabel("number of people")
plt.ylabel("payment method")
plt.xticks(rotation=45)
plt.legend(title="Gender")
```

## 5. Device Usage by Age Group: 
### Do younger and older age groups use different devices (iOS, Android, Web)? (Heatmap)


```python
plt.figure(figsize=(8,6))

df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 25, 40, 60, 100], labels=['<25', '26-40', '41-60', '60+'])
device_crosstab=pd.crosstab(df["AgeGroup"],df["Device"])
sns.heatmap(device_crosstab,cmap='YlGnBu',annot=True,fmt='d')
plt.title("Device Usage by Age Group")
plt.xlabel("Devices")
plt.ylabel("Age of group")
plt.show()
```

## 6. **Rating Distribution by Country**: 
###  Which country's customers have given the best `Rating` on average? (Bar Chart)


```python
df_orig=pd.read_csv("customer_clean_data.csv")

avg_rating = df_orig.groupby("Country")["Rating"].mean()
plt.figure(figsize=(8,6))
avg_rating.plot(kind='bar' )
plt.title("Average Rating by Country")
plt.xlabel("Country")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

## 7. **Order Frequency by Membership**: 
### Which membership tier places orders most frequently? (Violin Plot)


```python
plt.figure(figsize=(8,6))

sns.violinplot(data=df,x="Membership",y="Orders",palette='muted')
plt.title("Order Frequency by Membership")
plt.xlabel("Membership")
plt.ylabel("Orders")
plt.legend(title="Orders")
plt.xticks(rotation=45)
plt.show()
```

## 8. **Returned Items vs. Total Spent**: 
### Do customers who spend more return fewer items, or is there a different pattern? (Scatter Plot)


```python
plt.figure(figsize=(8,6))
plt.scatter(df["TotalSpent"],df["ReturnedItems"])
plt.grid(True,linestyle='--',alpha=0.7)
plt.xlabel("Total Spend")
plt.ylabel("Returned Items")
plt.show()
```

## 9. **Coupon Usage Impact**: 
### Does using `CouponUsed` (Yes/No) show a significant difference in `TotalSpent`? (Side-by-side Box Plot)


```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df["CouponUsed"],y=df["TotalSpent"],palette='viridis')
plt.title("Coupon Usage Impact")
plt.xlabel("CouponUsed")
plt.ylabel("Total Spend")
plt.legend(title=("CouponUsed"))
plt.show()
```
